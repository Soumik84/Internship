```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import selenium
from selenium import webdriver
```

```python
driver=webdriver.Chrome("E:/Data Trained/chromedriver_win32 (1)/chromedriver.exe")
```

```python
driver.get('https://www.olx.in/cars_c84')
```

```python
price=[]
year=[]
KM=[]
brand=[]
model=[]
place=[]

pricetag=driver.find_elements_by_xpath("//span[@class='_3GOwr']")
for i in pricetag:
    if i.text is None :
        price.append("--") 
    else:
        price.append(i.text.split(' ')[1])
yeartag=driver.find_elements_by_xpath("//div[@class='KFHpP']")
for i in yeartag:
    if i.text is None :
        year.append("--") 
    else:
        year.append(i.text.split('-')[0])
KMtag=driver.find_elements_by_xpath("//div[@class='KFHpP']")
for i in KMtag:
    if i.text is None :
        KM.append("--") 
    else:
        KM.append(i.text.split('-')[1].replace('km',''))
modeltag=driver.find_elements_by_xpath("//div[@class='_4aNdc']")
for i in modeltag:
    if i.text is None :
        model.append("--") 
    else:
        model.append(i.text)
brandtag=driver.find_elements_by_xpath("//div[@class='_4aNdc']")
for i in brandtag:
    if i.text is None :
        brand.append("--") 
    else:
        brand.append(i.text.split(' ')[0])
placetag=driver.find_elements_by_xpath("//div[@class='_1zvfB']")
for i in placetag:
    if i.text is None :
        place.append("--") 
    else:
        place.append(i.text.split('\n')[0])
```
