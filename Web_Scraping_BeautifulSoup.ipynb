# IMDB top 100 Movies

```python
page1=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
page1
```

```python
soup1=BeautifulSoup(page1.content)
soup1
```

```python
rating=[]
for i in soup1.find_all('div', class_="inline-block ratings-imdb-rating"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
Name=[]

for i in soup1.find_all("div",class_="lister-item mode-advanced"):
    Name.append(i.h3.a.text)
    
Name
```

```python
Year=[]
for i in soup1.find_all('span', class_="lister-item-year text-muted unbold"):
    Year.append(i.text.replace('(','').replace(')',''))
    
Year
```

```python
soup2=BeautifulSoup(page2.content)
soup2
```

```python
rating1=[]
for i in soup2.find_all('div', class_="inline-block ratings-imdb-rating"):
    rating1.append(i.text.replace('\n',''))
    
rating1
```

```python
Name1=[]
for i in soup2.find_all("div",class_="lister-item mode-advanced"):
    Name1.append(i.h3.a.text)
  
    
Name1
```

```python
Year1=[]
for i in soup2.find_all('span', class_="lister-item-year text-muted unbold"):
    Year1.append(i.text.replace('(','').replace(')',''))
    
Year1
```

```python
Final_Rating= rating+rating1
Final_Name= Name+Name1
Final_Year= Year+Year1
```

```python
print(len(Final_Rating),len(Final_Name),len(Final_Year))
```

```python
d=pd.DataFrame({'Name':Final_Name,'Year':Final_Year,'Rating':Final_Rating})
d
```
