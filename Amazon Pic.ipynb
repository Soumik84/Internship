```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get('https://www.amazon.in/s?k=saree+women&i=fashion&crid=37W9O6FN9QB02&sprefix=saree+wo%2Cfashion%2C254&ref=nb_sb_ss_ts-doa-p_1_8')
page
```

```python
s=BeautifulSoup(page.content)
s
```

```python
image=[]
for i in s.find_all('div',class_="a-section a-spacing-none _bXVsd_container_1DI1V _bXVsd_block_1vI8- _bXVsd_hFull_2lnNw _bXVsd_wFull_3f8b2 _bXVsd_row_3CEm0 _bXVsd_itemsCenter_3flfN _bXVsd_contentCenter_3BjNx"):
    image.append(i['src'])
          
image
```

```python
page=requests.get('https://www.amazon.in/s?k=trousers+for+men&i=fashion&crid=10U2TOCW27BH5&sprefix=trousers%2Cfashion%2C360&ref=nb_sb_ss_ts-doa-p_1_8')
page
```

```python
s1=BeautifulSoup(page.content)
s1
```

```python
image1=[]
for i in s1.find_all('img',class_="s-image"):
    image1.append(i['src'])
          
image1
```

```python
page=requests.get('https://www.amazon.in/s?k=jeans+for+men&i=fashion&crid=2ZS7L3TC2UK4F&sprefix=jeans+for+men%2Cfashion%2C251&ref=nb_sb_ss_ts-doa-p_1_13')
page
```

```python
s2=BeautifulSoup(page.content)
s2
```

```python
image2=[]
for i in s2.find_all('img',class_="s-image"):
    image2.append(i['src'])
          
image2
```

```python
Image=pd.DataFrame({})
Image['Trousers']=image1
Image['Jeans']=image2
```
